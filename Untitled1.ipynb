```r
data_with_outliers <- read.csv('/content/response.csv', header = TRUE, sep = ';')

Q <- quantile(data_with_outliers$custo, probs=c(.25, .75), na.rm = FALSE)

iqr <- IQR(data_with_outliers$custo)

data<- subset(data_with_outliers, data_with_outliers$custo > (Q[1] - 1.5*iqr) & data_with_outliers$custo < (Q[2]+1.5*iqr))

summary(data)
```

```r
summary(aov(custo ~ limite_pasteurizadora * tamanho_dos_estoques * ponto_do_estoque, data))
```
